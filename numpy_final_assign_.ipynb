
# Q10. Concatenation and Stacking

## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
arr1 = np.array([[2,5,7],[6,4,8]])
arr2 = np.array([[7,4,3],[4,7,2]])

print(np.vstack((arr1,arr2)))
print(np.concatenate((arr1,arr2),axis=1))
````
